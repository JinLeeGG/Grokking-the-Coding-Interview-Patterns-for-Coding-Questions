```python
class Solution(object):
    def isPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """
        s = s.lower()
        sl = []
        for i in range(len(s)):
            if s[i].isalnum():
              sl.append(s[i])

        if len(sl) == 1:
          return False

        left = 0
        right = len(sl) -1

        while left < right:
            if sl[left] != sl[right]:
                return False
            left += 1
            right -= 1

        return True


s = Solution()
s.isPalindrome("A man, a plan, a canal: Panama")
# s.isPalindrome("0P")
```
