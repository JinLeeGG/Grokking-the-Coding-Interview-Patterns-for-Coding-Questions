```python
# O(n log n)

class Solution(object):
    def sortedSquares(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        sqrtList = list()

        for i in range(len(nums)):
            sqrtList.append(nums[i] * nums[i])

        return sorted(sqrtList)

s = Solution()
s.sortedSquares([-7,-3,2,3,11])
```

```python
# O(n)
# Using Two Pointers
class Solution(object):
    def sortedSquares(self, nums):
        left = 0
        right = len(nums) - 1
        square = []

        while left <= right:
            if abs(nums[left]) > abs(nums[right]):
                square.append(nums[left] ** 2)
                left += 1
            else:
                square.append(nums[right] ** 2)
                right -= 1

        square.reverse()

        return square

s = Solution()
s.sortedSquares([-7,-3,2,3,11])
```
