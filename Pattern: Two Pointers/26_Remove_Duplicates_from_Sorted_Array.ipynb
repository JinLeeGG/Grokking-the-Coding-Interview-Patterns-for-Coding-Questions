```python
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        lengthOfNums = len(nums)
        remove_duplicate = list(set(nums))
        s = len(remove_duplicate)
        count_missingLength = lengthOfNums - len(remove_duplicate)

        for i in range(count_missingLength):
            remove_duplicate.append('_')

        return s, remove_duplicate

s = Solution()
s.removeDuplicates([0,0,1,1,1,2,2,3,3,4])
```

```python
# Use in Place O(n)
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # start at 1 because we need to compare it to previous value in nums
        left = 1

        # set right pointer in 1 as well.
        for right in range(1, len(nums)):
            # if previous value is differnt than current value in nums list (countered new unique value)
            if nums[right] != nums[right-1]:
                # set value at left pointer as right pointer value
                nums[left] = nums[right]
                left += 1

            # Otherwise the countered value is not a unique value

        return left

s = Solution()
s.removeDuplicates([0,0,1,1,1,2,2,3,3,4])
```
