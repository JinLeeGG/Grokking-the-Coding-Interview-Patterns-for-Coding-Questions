```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        count = 0
        hash = {}
        for num in nums:
           hash[num] = hash.get(num, 0) + 1 # 값이 없으면 0을 추가 한 후 1을 추가 / 있으면 1씩 증가
           count += hash[num] - 1 # 값이 2 이상이여야 카운트에 추가가 됨으로 -1. 1번만 나왓을경우, 0개의 페어 그리고 3번 나왔을경우 2개의 페어
        return count

s = Solution()
s.numIdenticalPairs([1,2,3,1,1,3])
```

```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        count = 0
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i] == nums[j]:
                    count += 1
        return count




s = Solution()
s.numIdenticalPairs([1,2,3,1,1,3])
```
