```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        count = 0
        hash = {}
        for num in nums:
           hash[num] = hash.get(num, 0) + 1 #   0    1  /  1 
           count += hash[num] - 1 #  2     -1. 1 , 0   3  2 
        return count

s = Solution()
s.numIdenticalPairs([1,2,3,1,1,3])
```

```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        count = 0
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i] == nums[j]:
                    count += 1
        return count




s = Solution()
s.numIdenticalPairs([1,2,3,1,1,3])
```
