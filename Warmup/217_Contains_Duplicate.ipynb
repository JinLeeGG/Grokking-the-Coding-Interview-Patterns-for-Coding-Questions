```python
def hasDuplicate(nums):
    if len(set(nums)) < len(nums):
      return False
    else:
      return True

hasDuplicate([1,2,3,4,3])
```

```python
class Solution:
    def containsDuplicate(self, nums):
      # TODO: Write your code here
      s1 = set(nums)
      if len(nums) is not len(s1):
          return True
      else:
          return False

      print(containsDuplicate([1,2,3,5]))
```
