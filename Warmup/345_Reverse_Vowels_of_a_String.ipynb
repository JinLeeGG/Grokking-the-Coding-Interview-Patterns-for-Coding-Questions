```python
class Solution(object):
    def reverseVowels(self, s):
        """
        :type s: str
        :rtype: str
        """
        data = list(s)
        vowelList = {'a','e','i','o','u','A','E','I','O','U'}
        left, right = 0, len(s)-1

        while(left < right):
            if data[left] not in vowelList:
                left += 1
            elif data[right] not in vowelList:
                right -= 1
            else:
                data[left], data[right] = data[right], data[left]
                left += 1
                right -= 1

        return ''.join(data)

s = Solution()
s.reverseVowels('hello')


```
