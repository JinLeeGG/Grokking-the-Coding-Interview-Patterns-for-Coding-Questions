```python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        sSet = sorted(list(s))
        tSet = sorted(list(t))

        print(sSet, tSet)
        count = 0

        for i in range(len(sSet)):
           if sSet[i] == tSet[i]:
              count+=1
           else:
              return False

        if count == len(sSet):
          return True



s = Solution()
s.isAnagram("anagram", "nagaram")
s.isAnagram("rat", "car")
```
