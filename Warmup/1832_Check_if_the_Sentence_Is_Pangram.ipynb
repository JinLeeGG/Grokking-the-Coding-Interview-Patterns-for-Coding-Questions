```python

def checkIfPangram(sentence):
  seen = set()
  for i in range(len(sentence)):
    seen.add(sentence[i])

  print(seen)

  if len(seen) == 27:
    return True
  else:
    return False

checkIfPangram("TheQuickBrownFoxJumpsOverTheLazyDog")
```
