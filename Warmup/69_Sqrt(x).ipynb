```python
class Solution(object):
    def mySqrt(self, x):
        """
        :type x: int
        :rtype: int
        """
        # Solve using binary search
        left = 0
        right = x
        mid = 0
        while left <= right:
          mid = (left + right) // 2
          if mid * mid < x:
            left = mid + 1
          elif mid * mid > x:
            right = mid - 1
          else:
            return mid

        return right

s = Solution()
s.mySqrt(9)

```
