```python
class Solution:
  def shortestDistance(self, words, word1, word2):
    # TODO: Write your code here
      pos1 = None
      pos2 = None
      minDist = float('inf')
      for i in range(len(words)):
          if words[i] == word1:
              pos1 = i
          elif words[i] == word2:
              pos2 = i

          if pos1 is not None and pos2 is not None:
              minDist = min(minDist, abs(pos1 - pos2))

      return minDist




s = Solution()
s.shortestDistance(["a", "c", "d", "b", "a"], 'a', 'b')
s.shortestDistance(["the", "quick", "brown", "fox", "jumps", "over", "the", "lazy", "dog"], word1 = "fox", word2 = "dog")


```
